**Key Topic: The `zip()` Function**

The `zip()` function allows you to iterate over multiple sequences simultaneously without needing a manual index variable.

**Basic Syntax:**
```python
for var1, var2 in zip(list1, list2):
    # code using var1 and var2
```

**Example with 2 lists:**
```python
movie_titles = ["The Umbrellas of Cherbourg", "M3gan", "Psycho"]
movie_ratings = [7.8, 6.3, 8.5]

for title, rating in zip(movie_titles, movie_ratings):
    print("The movie " + title + " has an IMDB rating of " + str(rating))
```

**Example with 3+ lists:**
```python
titles = ["The Umbrellas of Cherbourg", "M3gan", "Psycho"]
ratings = [7.8, 6.3, 8.5]
years = ["1964", "2022", "1960"]

for title, rating, year in zip(titles, ratings, years):
    print("'" + title + "' (" + year + ") has rating: " + str(rating))
```

**Important: `zip()` stops at the shortest sequence**
- If sequences have different lengths, `zip()` stops when the shortest sequence ends
- Example: `zip([1, 2, 3], [4, 5])` produces only 2 pairs: (1,4) and (2,5)

## Lecture 14: User-defined Functions

### What this lecture is about
You learned how to package repeated logic into functions using `def ...` so code is easier to reuse, test, and read.

### What the code means
- `def my_func(...):` creates a function.
- Inputs go inside parentheses (arguments).
- `return ...` sends output back.
- A function can return one value or multiple values.
- If you don’t return anything, it behaves like `print()`/plot helpers.

### Example
```python
def investment_value(P, r, n, t):
    # P = principal, r = annual rate (decimal), n = compounds/year, t = years
    return P * (1 + r/n)**(n*t)

V = investment_value(100, 0.01, 20, 10)
print(round(V, 2))  # about 110.51
```

Why this matters: instead of retyping the formula each time, you call one function with different inputs.

## Lecture 18b: Merging

### What this lecture is about
Combining related tables (files) into one DataFrame using shared keys like `raceId` and `circuitId`.

### What the code means
- `pd.merge(left=..., right=..., on="key", how="left")`
  matches rows where key values are equal.
- `how="left"` keeps all rows from the left table.
- Multiple merges can be chained to build one final dataset.

### Example
```python
import pandas as pd

df_results = pd.read_csv("f1_data/results.csv")
df_races = pd.read_csv("f1_data/races.csv")
df_circuits = pd.read_csv("f1_data/circuits.csv")

results_short = df_results[["resultId", "raceId", "driverId", "points"]]
races_short = df_races[["raceId", "year", "circuitId", "name"]]
circuits_short = df_circuits[["circuitId", "name", "location", "country"]]

merged1 = pd.merge(left=results_short, right=races_short, how="left", on="raceId")
merged_final = pd.merge(left=merged1, right=circuits_short, how="left", on="circuitId")

print(merged_final.head())
```

Why this matters: real projects almost always store data across multiple files/tables.

## Lecture 20b: Merging

### What this lecture is about
You expanded merge skills by combining multiple F1 tables, comparing merge directions (`left` vs `right`), and handling duplicate column names with suffixes.

### What the code means
- `pd.merge(left=..., right=..., on=..., how=...)` joins rows by shared keys.
- `how="left"` preserves row order/coverage of the left DataFrame.
- Swapping left/right changes output row behavior and column order.
- `suffixes=("_race", "_circuit")` resolves conflicting column names like `name`.
- Columns can be selected before merge to keep outputs focused.

### Example
```python
df_merged = pd.merge(
    left=df_results[["resultId", "raceId", "driverId", "points"]],
    right=df_races[["raceId", "year", "circuitId", "name"]],
    how="left",
    on="raceId"
)

df_merged_final = pd.merge(
    left=df_merged,
    right=df_circuits[["circuitId", "name", "location", "country"]],
    how="left",
    on="circuitId",
    suffixes=("_race", "_circuit")
)
```

Why this matters: most real datasets are relational and require careful key-based joins.

## Lecture 22b: Grouping and Aggregating

### What this lecture is about
You learned to compute group-level statistics using `groupby` and `agg`, chain operations, and merge aggregates back into row-level data.

### What the code means
- `groupby(...)` creates grouped views of a DataFrame.
- Selecting columns plus `.mean()` computes per-group averages.
- `.agg(...)` computes multiple stats with clear output column names.
- Chaining (`query().groupby().agg().sort_values()`) keeps workflows concise.
- Aggregated tables can be merged back for enriched row-wise analysis.

### Example
```python
drivers_agg = (df_results.groupby("driverId")
    .agg(mean_points=("points", "mean"),
         sd_points=("points", "std"),
         appearances=("points", len)))

results_merge = pd.merge(df_results, drivers_agg, on="driverId", how="left")
```

Why this matters: grouping + aggregating is essential for comparing performance across entities (drivers, teams, races, etc.).

### What this lecture is about
You reinforced group-level analysis workflows in Pandas by combining `groupby`, `agg`, filtering, sorting, and merge operations on Formula One race data.

### What the code means
- `df.groupby("key")` creates grouped views for per-entity calculations (e.g., driver, constructor, race).
- `grouped["col"].mean()` computes single-stat summaries, while `.agg(...)` computes multiple statistics at once.
- Chaining (`query().groupby().agg().sort_values()`) builds compact, readable workflows.
